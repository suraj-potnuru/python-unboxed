```python
# async - pause and resume
# await - tells python to wait untill an operation completes
```

```python
import time

def download_file(i):
    print("Downloading file", i)
    time.sleep(1)
    print("File downloaded")

def main():
    start = time.time()
    for i in range(1,4):
        download_file(i)
    end = time.time()
    print(f"Time taken is {end - start} seconds")

main()
```

```python
import time
import asyncio

async def download_file(i):
    print("Downloading file", i)
    await asyncio.sleep(1)
    print("File downloaded")

async def main():
    start = time.time()
    tasks = [download_file(i) for i in range(1,4)]
    await asyncio.gather(*tasks)
    end = time.time()
    print(f"Time taken is {end - start} seconds")

await main()
```
