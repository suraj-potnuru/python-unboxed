```python
from pydantic import BaseModel

class User(BaseModel):
    id: int
    name: str
    email: str
```

```python
user_1 = User(id=1, name="Suraj", email = "example@abc.com")
```

```python
user_2 = User(id="one", name="Suraj", email = "example@abc.com")
```

```python
user_3 = {
    "id": "soem string",
    "name": "John",
    "email": "abc@abc.com"
}
```

```python
user_3_validated = User(**user_3)
```
