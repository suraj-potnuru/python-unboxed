```python
def example_1():
    return 1
    return 2
    return 3
```

```python
example_1
```

```python
def example_2(): # this function becomes a generator
    print("Yielding 1")
    yield 1
    print("Yielding 2")
    yield 2
    print("Yielding 3")
    yield 3
```

```python
# next - next keyword can be used to access a generator

example_2()
```

```python
example_gen = example_2()
```

```python
next(example_gen)
```

```python
# Custom iterator can be implemented by any class that has these 2 methods

# __iter__ -> Creates the iterator object iteself
# __next__ -> implements the next method
```

```python
class CountUpTo:
    def __init__(self, limit):
        self.limit = limit
        self.count = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count <= self.limit:
            value = self.count
            self.count += 1
            return value
        else:
            raise StopIteration
```

```python
counter_iterator = CountUpTo(5)
```

```python
for num in counter_iterator:
    print(num)
```
