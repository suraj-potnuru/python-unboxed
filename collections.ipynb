```python
# Built in collections 

# lists, tuples, sets, dict

# Seperate module called collections (Counter, defaultdict, namedtuple etc)
```

```python
countries = ["India", "USA", "Japan", "Japan", "India", "India"]

count_dict = {}
for country in countries:
    if country in count_dict:
        count_dict[country] += 1
    else:
        count_dict[country] = 1
```

```python
count_dict
```

```python
from collections import Counter
```

```python
countries = ["India", "USA", "Japan", "Japan", "India", "India"]
count_dict_2 = Counter(countries)
```

```python
count_dict_2
```

```python
count_dict_2.most_common()[0][0]
```

```python
dir(count_dict_2)
```

```python
# ChainMap

from collections import ChainMap


defaults = {"theme": "Light", "lang": "en"}
user_settings = {"theme": "dark"}

config_cm = ChainMap(user_settings, defaults)
```
